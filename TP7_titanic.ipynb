```python
import pandas as pd
data=pd.read_csv("train.csv")
data.info()
```

```python
data.drop(['PassengerId','Cabin','Ticket','Name'],axis=1,inplace=True)
```

```python
data['FamilySize']=data['SibSp']+data['Parch']
data.drop(['SibSp','Parch'],axis=1,inplace=True)
```

```python
data.dropna(axis=0,how='any',inplace=True)
```

```python
data.head()
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
data['Sex']=encoder.fit_transform(data['Sex'])
data['Embarked']=encoder.fit_transform(data['Embarked'])
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
Y = data['Survived']
X = data.drop('Survived', axis=1)
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

# Model 1

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.neural_network import MLPClassifier
model = MLPClassifier(hidden_layer_sizes=(75,150,22),solver='sgd',activation='logistic')
model.fit(x_train,y_train)
from sklearn.model_selection import GridSearchCV
para = {'alpha':[0.000001,0.00001,0.001,0.01,0.0001],'hidden_layer_sizes':[(25,50),20,(150,100,200),(5,10)]}
model = GridSearchCV(model, para)
model.fit(x_train,y_train)
print("\n The best estimator :",model.best_estimator_)
print("\n the best score : ",model.best_score_)
```

# Model2

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.neural_network import MLPClassifier
model = MLPClassifier(hidden_layer_sizes=(75,150,22,100),solver='sgd',activation='relu')
model.fit(x_train,y_train)
from sklearn.model_selection import GridSearchCV
para = {'hidden_layer_sizes':[(25,50,40),20,(150,100,200,250),(5,10,30),(50,75,100,60)]}
model = GridSearchCV(model, para)
model.fit(x_train,y_train)
print("\n The best estimator :",model.best_estimator_)
print("\n the best score : ",model.best_score_)
```
